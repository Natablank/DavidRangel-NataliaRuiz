```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sym
import scipy.integrate as integrate
from IPython.display import IFrame
import time
```

```python
Q = np.array([0, 0.2,0.4,0.7,0.9,1])
N = 100
t = np.linspace(0,10,N)
```

```python
def GetRK4(q,c):
    U = np.zeros(len(t))
    U[0] = c
    for i in range(1,len(t)):
        h = 0.1
        k1 = U[i-1]**q
        k2 = (U[i-1]+0.5*k1*h)**q
        k3 = (U[i-1]+0.5*k2*h)**q
        k4 = (U[i-1]+k3*h)**q
        U[i] = U[i-1] + h/6*(k1+2*k2+2*k3+k4)
    return U
```

```python
U1 = GetRK4(Q[0],1)
U2 = GetRK4(Q[1],1)
U3 = GetRK4(Q[2],1)
U4 = GetRK4(Q[3],1)
U5 = GetRK4(Q[4],1)
U6 = GetRK4(Q[5],1)
```

```python
fig = plt.figure(figsize=(4,4))
ax = fig.add_subplot(1,1,1)
ax.plot(t,U1, color = '#438D80')
plt.xlabel('t(s)')
plt.ylabel('u(t)')
plt.title(f"$q={Q[0]}$")
plt.show()
```

```python
x = sym.Symbol('x',Real=True)
h = sym.Symbol('h',Real=True)

#1

f_0 = (x-(-h))*(x-(-2*h))/(2*h**2)
f_1 = (x-(-2*h))*x/(-h**2)
f_2 = (x-(-h))*x/(2*h**2)

b_0 = sym.integrate(f_0,(x,0,h))
b_1 = sym.integrate(f_1,(x,0,h))
b_2 = sym.integrate(f_2,(x,0,h))

print(f'Los coeficientes del método de Adams-Bashforth de 3 puntos son: b_0={b_0}, b_1={b_1}, b_2={b_2}')

#2

f_0 = (x-(-h))*(x-(-2*h))*(x-(-3*h))/(6*h**3)
f_1 = (x-(-2*h))*(x-(-3*h))*x/(-2*h**3)
f_2 = (x-(-h))*(x-(-3*h))*x/(2*h**3)
f_3 = (x-(-h))*(x-(-2*h))*x/(-6*h**3)

b_0 = sym.integrate(f_0,(x,0,h))
b_1 = sym.integrate(f_1,(x,0,h))
b_2 = sym.integrate(f_2,(x,0,h))
b_3 = sym.integrate(f_3,(x,0,h))

print(f'Los coeficientes del método de Adams-Bashforth de 4 puntos son: b_0={b_0}, b_1={b_1}, b_2={b_2}, b_3={b_3}')

#3

f_0 = x*(x-(-h))/(2*h**2)
f_1 = (x-(-h))*(x-h)/(-1*h**2)
f_2 = (x-h)*x/(2*h**2)

b_0 = sym.integrate(f_0,(x,0,h))
b_1 = sym.integrate(f_1,(x,0,h))
b_2 = sym.integrate(f_2,(x,0,h))

print(f'Los coeficientes del método de Adams-Moulton de 3 puntos son: b_0={b_0}, b_1={b_1}, b_2={b_2}')

#4

f_0 = (x-(-h))*(x-(-2*h))*x/(6*h**3)
f_1 = (x-(-h))*(x-h)*(x-(-2*h))/(-2*h**3)
f_2 = (x-h)*(x-(-2*h))*x/(2*h**3)
f_3 = (x-h)*(x-(-h))*x/(-6*h**3)

b_0 = sym.integrate(f_0,(x,0,h))
b_1 = sym.integrate(f_1,(x,0,h))
b_2 = sym.integrate(f_2,(x,0,h))
b_3 = sym.integrate(f_3,(x,0,h))

print(f'Los coeficientes del método de Adams-Moulton de 4 puntos son: b_0={b_0}, b_1={b_1}, b_2={b_2}, b_3={b_3}') 
```
