```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sym
import scipy as scy
from scipy.integrate import solve_ivp
from IPython.display import display, Math, Latex
from termcolor import colored
```

```python
Q = np.array([0., 0.2, 0.4, 0.7, 0.9, 1.])
t = np.linspace(0, 10, 30)
u0 = 0
r0 = np.array([0, 1])
```

```python
def solucion():
  s = np.zeros(len(Q))
  for i in range(len(Q)):
    q = Q[i]
    def ecuacion(q_, u_):
      if q == 1:
        du = (u_)
      else:
        du = u_**q 
      return du

    sl = solve_ivp(ecuacion, (t[0], t[-1]), r0, t_eval = t, method = 'RK45')
    s = np.append(s, sl)
  return s
```

```python
sl = solucion()
print(colored('Punto 1. Ecuación diferencial no lineal', attrs = ['bold']))
fig, ax = plt.subplots(3, 2)

c = 0 
c_ = 6
for i in range(3):
  for j in range(2):
    ax[i, j].set_title(f'solución = {Q[c]}')
    ax[i, j].plot(t, sl[c_].y[1], marker = 'o', label = 'sol. numérica', color = '#800000')
    if c == 5:
      ax[i, j].plot(t, np.e**t, marker = '.', label = 'sol. analítica', color = '#9F000F')
    else: 
      ax[i, j].plot(t, (t*(1-Q[c])+1)**(1/(1-Q[c])), marker = '.', label = 'sol. analítica', color = '#571B7E')
    ax[i, j].legend()
    c += 1
    c_ += 1
plt.show
```
