```python
#Con lo visto en clase:
def GetRelaxation(T, Nit = int(1e5), omega = 1.9, tolerancia = 1e-2):
    
    itmax = 0
    
    for it in tqdm(range(Nit)):
        
        dmax = 0.
        
        for i in range(1, len(x)-1):
            for j in range(1, len(y)-1):
                tmp = 0.25*(T[i+1,j]+T[i-1,j]+T[i,j+1]+T[i,j-1])
                r = omega*(tmp - T[i,j])
                
                T[i,j] += r
                
                if np.abs(r) > dmax:
                    dmax = r
       # print(T)
       # print(it)
        
        if np.abs(dmax) < tolerancia:
            print(it)
            itmax = it
            break
            
    return T,itmax
```

```python
def GetGrad(T):
    
    M,N = T.shape
    
    u_ = np.zeros_like(T)
    v_ = np.zeros_like(T)
    
    for i in range(M):
        for j in range(N):
            
            if j == 0:
                v_[i,j] = T[i,j+1] - T[i,j]
            elif j == N - 1:
                v_[i,j] = T[i,j] - T[i,j-1]
            else:
                v_[i,j] = (T[i,j+1] - T[i,j-1])/2.
                
                
            if i == 0:
                u_[i,j] = T[i+1,j] - T[i,j]
            elif i == M - 1:
                u_[i,j] = T[i,j] - T[i-1,j]
            else:
                u_[i,j] = (T[i+1,j] - T[i-1,j])/2.
                
                
    return u_,v_
```
