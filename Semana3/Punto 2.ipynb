```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

##2.

###a)


```python
dt = pd.read_csv('https://raw.githubusercontent.com/asegura4488/Database/main/MetodosComputacionalesReforma/EnergiaPotencialGas2D.txt')
```

```python
datos = dt.iloc[:,0].str.split(expand=True)
```

```python
tiempo = pd.to_numeric(datos.iloc[:,0])
Energia_potencial = pd.to_numeric(datos.iloc[:,1])
```

###b)


```python
valor_medio = np.mean(Energia_potencial)
for i in range(len(tiempo)):
  if np.round(tiempo[i],2) == np.round(valor_medio,2):
    tiempo[i].delete()
    Energia_potencial[i].delete()
```

###c)

```python
Transformada_Fourier = np.fft.fft(Energia_potencial)
```

###d)

```python
Frecuencias = np.fft.fftfreq(Energia_potencial.size,(np.max(tiempo)-np.min(tiempo)/(tiempo.size)))
F_fundamental = np.max(Frecuencias)
```

###e)

```python
T = 1/F_fundamental
```

###f)

```python
x = tiempo

y = 10*np.cos(100*F_fundamental*x)


fig = plt.figure(figsize=(9,6))
plt.plot(tiempo,Energia_potencial,'--',label='Energ√≠a potencial')
plt.xlabel('t(s)')
plt.ylabel('E[U]')
plt.legend()


fig = plt.figure(figsize=(9,6))
plt.plot(tiempo,Energia_potencial,'--',label='Serie original', c='blue')
plt.plot(x,y,label='Fundamental',c='red')
plt.xlabel('t(s)')
plt.ylabel('E[U]')
plt.legend()
```
